```python
N=53
eps=np.float64(1.0)

for k in range(N):
    eps=eps/2
    pluseps= 1.0+eps
    error= pluseps-eps
    print('eps= ', eps, ', one + eps = ',pluseps)
    print('Error= ', error)
    print('loop= ', k)
    
    

epsc=np.complex(1)

for k in range(N):
    epsc=epsc/2
    plusepsc= 1.0+epsc
    errorc= plusepsc-epsc
    print('epsc= ', epsc, ', one + eps = ',plusepsc)
    print ('Error= ', errorc)
    print('loop= ', k)
```

For Double precision floats, the precision is 2.220446049250313e-16. Anything smaller than that and the computer can't keep track.

For complex numbers, the precision is 2.220446049250313e-16+0j. Anything smaller than that and the computer can't keep track.
